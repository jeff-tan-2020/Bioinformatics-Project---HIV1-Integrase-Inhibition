```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyRegressor
```

## 1. Load data with independent (x) and dependent (y) variables

```python
df = pd.read_csv('HIV1-inhibitor_pIC50_pubchem_fp.csv')
```

```python
X = df.drop('pIC50', axis=1)
Y = df.pIC50
```

## 2. Data Preprocessing & Splitting data into test (70) : training (30)

```python
# Examine X dimension
X.shape
```

```python
# Remove low variance features
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))    
X = selection.fit_transform(X)
X.shape
```

```python
# Perform data splitting using 70/30 ratio
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2021)
```

## 3. Compare ML algorithms

```python
# Defines and builds the lazyclassifier
clf = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, Y_train, Y_test)
```
